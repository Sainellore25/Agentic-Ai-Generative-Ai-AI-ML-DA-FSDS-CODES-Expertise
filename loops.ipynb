```python
i = 1

while i<5:
    print('data science')  # when we mention end then new line will not create 
    j = 1
    while j<=4:
        print('technology')
        j = j + 1
        
    i = i + 1
    print()
    
  # The output  which we got is very lengthy but how to make them one line lets refer to  below code
```

```python
i = 1
while i<=5:
    print('data science', end = "") # when we mention end then new line will not create
    j = 1
    while j<=4:
        print('technology',end ="")
        j = j + 1
    i = i + 1
    print() 
```

```python
# lets use while loop using some numbers
i = 1

while i <= 2 :
    j = 0
    while j <= 2 :
        print(i*j, end="")
        j += 1
    print()
    i += 1
```

```python
# lets use while loop using some numbers
i = 1
while i <= 4 :
    j = 0
    while j <= 3 :
        print(i*j, end="")
        j += 1
    print()
    i += 1
```
