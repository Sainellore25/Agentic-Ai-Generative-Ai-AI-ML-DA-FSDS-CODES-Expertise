```python
from deep_translator import GoogleTranslator
from gtts import gTTS
from IPython.display import Audio

# English text
text = """I am going to China. I met Yang there. 
I was very happy when I saw him. 
I don't know why, but my heart rate was very high. 
Then I realized it was not a dream, it was real."""

# Step 1: Translate to Hindi
hindi_text = GoogleTranslator(source='auto', target='hi').translate(text)
print(hindi_text)

# Step 2: Convert to speech
tts = gTTS(hindi_text, lang='hi')
filename = "hindi_speech.mp3"
tts.save(filename)

# Step 3: Play sound inside Jupyter Notebook
Audio(filename, autoplay=True)
```

```python
import os
os.startfile("hindi_speech.mp3")
```

```python
from deep_translator import GoogleTranslator
from gtts import gTTS
from IPython.display import Audio

# English text
text = """I am going to China. I met Yang there. 
I was very happy when I saw him. 
I don't know why, but my heart rate was very high. 
Then I realized it was not a dream, it was real."""

# Step 1: Translate to Korean
korean_text = GoogleTranslator(source='auto', target='ko').translate(text)
print("Translated Korean Text:")
print(korean_text)

# Step 2: Convert Korean text to speech
tts = gTTS(text=korean_text, lang='ko')
filename = "korean_speech.mp3"
tts.save(filename)

# Step 3: Play sound in Jupyter Notebook
Audio(filename, autoplay=True)
```
