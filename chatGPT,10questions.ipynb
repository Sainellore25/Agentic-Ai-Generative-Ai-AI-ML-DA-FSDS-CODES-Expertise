# 1.Write a Function to find the Square of a Number.

```python
def square(num):
    return num*num

print(square(5)) 
```

# 2. Write a function to check if a number is even or odd.

```python
def even_odd(n):
    if n%2 == 0:
        return "Even"
    else:
        return "Odd"
print(even_odd(7)) 
```

# 3. Write a function to return the largest of three numbers.

```python
def largest(a,b,c):
    return max(a,b,c)

print(largest(10,25,18)) 
```

# 4.Write  a function to calculate factorial of a number.

```python
def factorial(n):
    fact = 1
    for i in range(1,n+1):
        fact *= i
    return fact

print(factorial(5))
```

# 5.Write a function to count vowels a string.

```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for ch in text:
        if ch in vowels:
           count += 1
    return count

print(count_vowels("Hello World"))     
```

# 6.Write a function to check if a string is a palindrome.

```python
def is_palindrome(s):
    return s == s[::-1]

print(is_palindrome("madam"))    
```

# 7.Write a function to return the sum of elements in a list.

```python
def list_sum(lst):
    return sum(lst)

print(list_sum([1,2,3,4]))
```

# 8 Write a function to find the Fibonacci series up to n terms.

```python
def fibonacci(n):
    result = []
    a,b = 0,1
    for _ in range(n):
        result.append(a)
        a,b = b,a+b
    return result

print(fibonacci(7))
```

# 9. Write a function to check if a number is prime.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

print(is_prime(13))
```

# 10.Write a function that takes a list & returns another list with only even numbers.

```python
def filter_even(lst):
    return[x for x in lst if x % 2 == 0]

print(filter_even([1,2,3,4,5,6]))
```
